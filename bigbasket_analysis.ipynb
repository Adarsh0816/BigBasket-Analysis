# Find out the outliers from the dataset according to the columns and fill them with the mean

```python
q1 = df['sale_price'].quantile(0.25)
q3 = df['sale_price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
df[(df['sale_price'] < lower_bound) | (df['sale_price'] > upper_bound)]["sale_price"]= df.sale_price.median()
```

```python
mean_value = df['sale_price'].mean()
df['sale_price'] = df['sale_price'].apply(
    lambda x: mean_value if x < lower_bound or x > upper_bound else x
)
print(mean_value)
```

```python
median_value = df['sale_price'].median()
df ['sale_price'] = df ['sale_price'].apply(
    lambda x: median_value if x < lower_bound or x > upper_bound else x
)
print(median_value)
```

```python
# boxplot of sales price of after replace with median
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['sale_price'], color='lightblue')
plt.title(" Aftter outliers replace with median ")
plt.xlabel("Sale Price")
plt.show()
```

# Create Plots or visualizations.

```python
# Histogram for Sale Price
plt.figure(figsize=(10, 6))   #   the size of the figure. figsize=(10, 6) means the plot will be 10 units wide and 6 units high.
sns.histplot(df['sale_price'], kde=True, color='blue')   # This is a seaborn function that creates a histogram using the sale price data. 
#It means that the Kernel Density Estimate (KDE) will also be plotted, which creates a smooth curve of the data.
plt.title('Distribution of Sale Price')   # This code  the title of the plot, which is "Distribution of Sale Price".
plt.xlabel('Sale Price')  #  The label of the X-axis is set to "Sale Price."
plt.ylabel('Frequency')  # The label of the y-axis is set to "'Frequency'."
plt.show()  # This code displays the plot.
```

```python
# Scatter Plot between Sale Price and Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sale_price'], y=df['rating'], color='purple')
plt.title('Sale Price vs Rating')
plt.xlabel('Sale Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Bar Plot for Product Categories
plt.figure(figsize=(12, 8))
category = df['category'].value_counts()
sns.barplot(x=category.index, y=category.values, palette='viridis')
plt.title('Product Categories Frequency')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
